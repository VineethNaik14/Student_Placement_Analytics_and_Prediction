# Exploratory Data Analysis (EDA)

## Summary

1. Aptitude Test Score shows the largest difference between placed and non-placed students.

   * Average difference: 11.29 points.

2. Coding Skills shows a strong positive relationship with placement.

   * Average difference: 1.82 points.

3. Communication Skills is positively associated with placement.

   * Average difference: 1.03 points.

4. CGPA is positively associated with placement.

   * Average difference: 1.02 points.

5. Students with more projects tend to have better placement outcomes.

   * Average difference: 0.96 projects.

6. Students with more certifications tend to have better placement outcomes.

   * Average difference: 0.64 certifications.

7. Students with more internships tend to have better placement outcomes.

   * Average difference: 0.54 internships.

8. Backlogs show a strong negative relationship with placement.

   * Placed students have significantly fewer backlogs on average.

9. Soft Skills Rating shows little to no relationship with placement.

   * Average ratings are nearly identical for placed and non-placed students.

10. Correlation analysis identified Projects (0.50), CGPA (0.49), Certifications (0.47), and Coding Skills (0.44) as the strongest positive predictors of placement.

11. Backlogs showed the strongest negative correlation with placement (-0.49), indicating that students with more backlogs are less likely to be placed.

12. Age (0.00) and Soft Skills Rating (-0.00) showed almost no correlation with placement outcomes.

13. Strong relationships were observed between several independent features:

    * Projects and Coding Skills (0.90)
    * Coding Skills and Certifications (0.86)
    * Projects and Certifications (0.83)
    * CGPA and Projects (0.71)
    * CGPA and Aptitude Test Score (0.71)

14. These relationships suggest that academic performance, practical experience, technical skills, and certifications collectively contribute to placement success.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/train.csv")
```

```python
# CGPA Analysis
df.groupby("Placement_Status")["CGPA"].mean()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data = df, x = "Placement_Status", y = "CGPA")
plt.title("Avg CGPA by Placement status")
plt.show()
```

```python
df["Placement_Status_Num"] = df["Placement_Status"].map({"Placed" : 1, "Not Placed" : 0})
```

```python
df[["Placement_Status", "Placement_Status_Num"]].head()
```

```python
numeric_cols = [
    "Age",
    "CGPA",
    "Internships",
    "Projects",
    "Coding_Skills",
    "Communication_Skills",
    "Aptitude_Test_Score",
    "Soft_Skills_Rating",
    "Certifications",
    "Backlogs",
    "Placement_Status_Num",
]

df[numeric_cols].corr()["Placement_Status_Num"].sort_values(ascending=False)
```

```python
corr_matrix = df[numeric_cols].corr()
```

```python
plt.figure(figsize=(10, 8))

sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
