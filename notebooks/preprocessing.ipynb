# Data Preprocessing

## Summary

1. Student_ID was removed because it is an identifier and does not contribute to placement prediction.

2. Features (X) and target variable (y) were separated.

   * X contains all predictor variables.
   * y contains the Placement_Status which is our target variable.

3. Gender was label encoded.

   * Male - 0
   * Female - 1

4. Degree was transformed using One-Hot Encoding.

   * Degree_BCA
   * Degree_B.Sc
   * Degree_B.Tech
   * Degree_MCA

5. Branch was transformed using One-Hot Encoding.

   * Branch_CSE
   * Branch_Civil
   * Branch_ECE
   * Branch_IT
   * Branch_ME

6. Original categorical columns (Degree and Branch) were replaced by their encoded representations.

7. The final dataset contains only numerical features suitable for machine learning algorithms.

8. Final feature matrix shape:

   * Rows: 45,000
   * Columns: 20

9. No missing values were present in the dataset.

10. All preprocessing steps were completed successfully and the dataset is ready for model training.


```python
import pandas as pd

df = pd.read_csv("../data/train.csv")
```

```python
X = df.drop("Placement_Status", axis=1)
y = df["Placement_Status"]
X.head()
y.head()
```

```python
X = X.drop("Student_ID",axis=1)
```

```python
X["Gender"] = X["Gender"].map({"Male" : 0, "Female" : 1})
```

```python
X = pd.get_dummies(X, columns=["Degree", "Branch"], dtype=int)
```
